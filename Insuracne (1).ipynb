```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
df=pd.read_csv("insurance_claims.csv")
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
col_to_del=["policy_number","policy_bind_date","insured_zip","insured_hobbies","_c39","insured_relationship","incident_date","auto_model","auto_year"]
df.drop(col_to_del,axis="columns",inplace=True)
df.head()
```

```python
X=df.drop("fraud_reported",axis="columns")
X.head()
```

```python
Y=df.fraud_reported
Y.head()
```

```python
dummy=pd.get_dummies(X)
dummy.head()
```

```python
dummy.shape
```

```python
dummy.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
x_scaled=scalar.fit_transform(dummy)
```

```python
Y.unique()
```

```python
Y.replace({"Y":0,"N":1},inplace=True)
Y.unique()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x_scaled,Y,test_size=0.2,random_state=0,stratify=Y)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_train.dtype
```

```python
from keras.api._v2.keras import activations

model=keras.Sequential([
    keras.layers.Dense(500,input_dim=1095,activation="relu"),
    keras.layers.Dropout(0.75),

    keras.layers.Dense(200,activation="relu"),
    keras.layers.Dropout(0.75),

    keras.layers.Dense(150,activation="relu"),
    keras.layers.Dropout(0.75),

    keras.layers.Dense(100,activation="relu"),
    keras.layers.Dropout(0.55),

    keras.layers.Dense(50,activation="relu"),
    keras.layers.Dropout(0.5),


    keras.layers.Dense(1,activation="sigmoid")
])
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.fit(x_train,y_train,epochs=100)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred=model.predict(x_test)
```
